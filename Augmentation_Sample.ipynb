```python
from tensorflow.keras.preprocessing.image import load_img
```

```python
img = load_img('C:/Augment/Bengin case (1).jpg',target_size=(200,200))
```

```python
type(img)
```

```python
img
```

```python
datagen = ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,    
    fill_mode='nearest'
)
```

```python
from tensorflow.keras.preprocessing.image import img_to_array
```

```python
img = img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
i=0
```

```python
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='C:/aug'):
    i=i+1
    if i==10:
        break
```
